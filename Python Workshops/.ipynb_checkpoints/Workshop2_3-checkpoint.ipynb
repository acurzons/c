```python
#function to remove conjoining lines between contours
def plot_contour(filename, image_header, color='white', ls='--', lw=2, alpha = 1.0, zorder=0):
    
    con = np.loadtxt(filename)
    
    wcs = WCS(image_header)
    
    uu = []
    
    for c in range(len(con)):
        s = SkyCoord(con[c][0], con[c][1], frame='galactic', unit='deg')
        uu.append(utils.skycoord_to_pixel(s,wcs))
        
    con = np.array(uu).T   
    
    i = 0
    while i < len(con.T):
        j = i
        i = np.argwhere(con[0]==con[0][i])[1][0] + 1
        plt.plot(con[0][j:i],con[1][j:i], color = color, ls = ls, lw =lw, alpha=alpha, zorder = zorder)
    
    return uu
    
```

```python
fig = plt.figure()


wcs = WCS(image_header)
ax = WCSAxes(fig, [0,0,1,1], wcs= wcs)
fig.add_axes(ax)

lat = ax.coords['glat']
lon = ax.coords['glon']

#format axes
lat.set_major_formatter('d.dd')
lon.set_major_formatter('d.dd')

plt.xlabel('Galactic Longitude / deg')
plt.ylabel('Galactic Latitude / deg')

plt.imshow(image_data, cmap=plt.cm.afmhot_r, vmin=low, vmax=high, zorder = 0)
cbar = plt.colorbar()
cbar.set_label('SomeLabel', rotation=270, labelpad=20)
#labelpad moves the label to the right

plt.xlim(99,633)
plt.ylim(35,581)

plot_contour('Archive/contours1.con',image_header, color='k', ls = '-', alpha = 0.7, zorder=1)
# plt.savefig('workshop2image.png')

#Add regions
reg = pyregion.open('Archive/C.reg')
r = reg.as_imagecoord(image_header)
patch_list, artist_list = r.get_mpl_patches_texts()

for p in patch_list:
    p.zorder = 2
    p.set_edgecolor('white')
    p.set_facecolor('white')
    p.set_alpha(0.7)
    ax.add_patch(p)
    
for t in artist_list:
    t.zorder = 2
    t.set_color("green")
    ax.add_artist(t)

plt.show()
```
